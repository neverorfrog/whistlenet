```python
import warnings
warnings.filterwarnings("ignore")
from utils import project_root

root = project_root()
```

```python
from models import MNIST_DATA_CONFIG
from data.mnist import MnistDataset
dataset = MnistDataset(tobeloaded=True, params=MNIST_DATA_CONFIG)
```

```python
from models.cnn import CNN

name = "mnist"
model = CNN(name)
```

```python
from models import MNIST_TRAIN_CONFIG
import time
from core.trainer import Trainer
train_model = False

trainer = Trainer(params=MNIST_TRAIN_CONFIG)
start_time = time.time()
if train_model:
    trainer.fit(model,dataset)
    model.training_time = time.time() - start_time
```

```python
import matplotlib.pyplot as plt
complete_plot = False

model.load(name)
plt.plot(model.test_scores, label=f'{name} - test scores')
if complete_plot:
    plt.plot(model.train_scores, label=f'{name} - train scores')
    plt.plot(model.val_scores, label=f'{name} - val scores')
    
plt.legend()
plt.ylabel("score")
plt.xlabel("epoch")
plt.show()
```
